```
storm = pd.read_csv('Storm15-18.txt', parse_dates=['Date'])
mystics = pd.read_csv('Mystics15-18.txt', parse_dates=['Date'])
storm.head()
```

```
#Rename Unnamed: 2 as HI for home indicator. Unnamed: 4 is WI for Win indicator
mystics.rename(columns={'Unnamed: 2': 'HI', 'Unnamed: 4': 'WI'}, inplace=True)
mystics['Visitor'] = np.where(mystics['HI']!='@', mystics['Opponent'], 'Washington Mystics')
mystics['Home'] = np.where(mystics['HI']=='@', mystics['Opponent'], 'Washington Mystics')
mystics['PTS_Visitor'] = np.where(mystics['HI']!='@', mystics['Opp'], mystics['Tm'])
mystics['PTS_Home'] = np.where(mystics['HI']=='@',  mystics['Opp'], mystics['Tm'])
```

```
mystics.head(8)
```

```
#Rename Unnamed: 2 as HI for home indicator. Unnamed: 4 is WI for Win indicator
storm.rename(columns={'Unnamed: 2': 'HI', 'Unnamed: 4': 'WI'}, inplace=True)
storm['Visitor'] = np.where(storm['HI']!='@', storm['Opponent'], 'Seattle Storm')
storm['Home'] = np.where(storm['HI']=='@', storm['Opponent'], 'Seattle Storm')
storm['PTS_Visitor'] = np.where(storm['HI']!='@', storm['Opp'], storm['Tm'])
storm['PTS_Home'] = np.where(storm['HI']=='@',  storm['Opp'], storm['Tm'])
storm.head()
```

```
df = pd.concat([storm, mystics])
df.head()
```

```
df = df[['Date', 'Visitor', 'PTS_Visitor', 'Home', 'PTS_Home']]
```

```
#Establish scoring method
from sklearn.metrics import f1_score, make_scorer, classification_report

scorer = make_scorer(f1_score, pos_label=None, average = 'weighted')
```

```
df['HomeWin'] = np.where(df['PTS_Visitor']< df['PTS_Home'], 1, 0)
df.head()
```

```
#Target Variable. Did hometeam win, Yes or No? 0 is no, 1 is yes
y_yes = df['HomeWin'].values
```

```
#Baseline measure. How often did the hometeam win Mystics and Storm 2015 through 2018?
y_yes.mean()
```

```
from sklearn.metrics import f1_score

#What happens if we always predict that the home team will win?
y_pred = [1] * len(y_yes)
f1 = f1_score(y_yes, y_pred, pos_label=None, average = 'weighted')
print(f1)
```

```
print(len(y_yes) - len(y_pred))
```

```
df['PTS_Visitor'].replace(to_replace='Forfeit', value=0, inplace=True)
df['PTS_Home'].replace(to_replace='Forfeit', value=0, inplace=True)
```

```
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier(random_state = 42)
from sklearn.cross_validation import cross_val_score
```

```
x=df[['PTS_Visitor', 'PTS_Home']].values
```

```
scores = cross_val_score(clf, x, y_yes, scoring=scorer)
np.mean(scores)
```
